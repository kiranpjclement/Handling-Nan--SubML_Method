## Filling Nan Age Using the Model to predict Age

```python
filt=df['Age'].isnull()
Null_Age_df=df.loc[filt]
Null_Age_df
```

```python
X_Null_Age_df=Null_Age_df[['Pclass','Sex','SibSp','Fare','Embarked','Survived']]
```

```python
temp_x1=pd.get_dummies(X_Null_Age_df['Embarked'],drop_first=True)
X_Null_Age_df=pd.concat([X_Null_Age_df,temp_x1],axis=1)
X_Null_Age_df.drop(columns='Embarked',inplace=True)
```

```python
temp_x2=pd.get_dummies(X_Null_Age_df['Sex'],drop_first=True)
X_Null_Age_df=pd.concat([X_Null_Age_df,temp_x2],axis=1)
X_Null_Age_df.drop(columns='Sex',inplace=True)
```

```python
temp_x3=pd.get_dummies(X_Null_Age_df['Survived'],drop_first=True)
X_Null_Age_df=pd.concat([X_Null_Age_df,temp_x3],axis=1)
X_Null_Age_df.drop(columns='Survived',inplace=True)
```

```python
X_Null_Age_df
```

#### Calling the Joblib file to reuse the model

```python
from joblib import dump, load
mod3 = load('Age_prediction.joblib')  #load the data in another notebook
pred=mod3.predict(X_Null_Age_df)
```
